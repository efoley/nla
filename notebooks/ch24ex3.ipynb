```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

from nla.plot import *
from nla.spect import *
```

```python
gen = np.random.default_rng(seed=87)

# note that by Gershgorin circle theorem this should mostly produce matrices
# with spectral abscissa less than 1
A = gen.normal(size=(10, 10)) - 2*np.eye(10)
spectral_abscissa(A)
```

```python
t = np.linspace(0, 20, 101)

mm_norms = np.linalg.norm(scipy.linalg.expm(t[:, None, None]*A), axis=(1,2), ord=2)
```

```python
plot_eigenvalues(A)

# Add a note below the plot
plt.gcf().text(0.5, 0, f"Spectral abscissa: {spectral_abscissa(A):0.3f}", ha='center', fontsize=10)

# Adjust layout to make room for the note
plt.tight_layout(rect=[0, 0.1, 1, 1])  # Adjust bottom space for text
```

```python
# The norm of the exp matrix will grow/shrink if the spectral abscissa is
# >1/<1 respectively. 
plt.plot(t, np.exp(t*spectral_abscissa(A)), label='exp spectral abscissa')
plt.plot(t, mm_norms, label='norm2 of exp matrix')
plt.legend()
```
