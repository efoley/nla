```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

from nla.random import random_tridiagonal
from nla.plot import plot_matrix
```

```python
gen = np.random.default_rng(seed=1245)

# make a symmetric tridiagonal matrix
A = random_tridiagonal(32, gen)
A = A + A.T

plt.matshow(A)
```

```python
Q, R = np.linalg.qr(A)

fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].set_title("Q")
plot_matrix(Q != 0, ax[0])
ax[1].set_title("R")
plot_matrix(R !=0 , ax[1])
```

```python
plot_matrix(R @ Q)
```
